{ **Performing Data cleansing before doing Analysis** }

```python
df['Business Tax Rate'] = df['Business Tax Rate'].apply(str)
df['Business Tax Rate'] = df['Business Tax Rate'].str.replace('%','').str.replace(',','').astype(float)
df['GDP'] = df['GDP'].apply(str)
df['GDP'] = df['GDP'].str.replace('$','').str.replace(',','').astype(float)
df['Health Exp/Capita'] = df['Health Exp/Capita'].apply(str)
df['Health Exp/Capita'] = df['Health Exp/Capita'].str.replace('$', '').str.replace(',', '').astype(float)
df
```

```python
df.isna().sum()
```

```python
# Removing columns with a high number of missing values
df.drop('Energy Usage', inplace=True, axis=1)
df.drop('Lending Interest', inplace=True, axis=1)
```

{ **Handling Missing Data** }

```python
# Applying the Imputation method to handle missing values and replace them with the column mean
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values= np.NaN, strategy='mean',verbose=0)
imputed_Df = imp.fit_transform(df.iloc[:, 0:16])
imputed_Df1 = pd.DataFrame(imputed_Df, columns = df.iloc[:,0:16].columns)
imputed_Df1
```

{ **Performing Scaling on the Data** }

```python
# Implementing StandardScaler for Scaling and importing StandardScaler from the Sklearn library
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
scaled = std.fit_transform(col_df)
df2 = pd.DataFrame(scaled, columns= imputed_Df1.iloc[:,[9,10,15,3]].columns) 
df2
```

```python
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
 
df_minmax_scal = imputed_Df1[['Life Expectancy Female','Life Expectancy Male', 'Population Urban','GDP']]
 
scaler = MinMaxScaler()
 
df2 = scaler.fit_transform(df_minmax_scal.to_numpy())
df2 = pd.DataFrame(df2, columns=['Life Expectancy Female','Life Expectancy Male', 'Population Urban','GDP'])
df2
```



**1) Kmeans Clustering**

```python
# Performing sample Kmeans clustering with taking cluster value =7
from sklearn.cluster import KMeans
model = KMeans(n_clusters=7)
model.fit(df2.iloc[:,0:])
print('The SSE Value of clusters 7 is: {:.2f}'.format(model.inertia_))
```

```python
#Determing SSE values for different clusters. 
df_final = pd.DataFrame(columns= ['Num_of_clusters', 'Inertia_Value'])
KS = range(1,11)
for k in KS:
    model = KMeans(n_clusters=k)
    model.fit(df2.iloc[:,0:])
    model.inertia_
    df_final = df_final.append({'Num_of_clusters': k, 'Inertia_Value': model.inertia_}, ignore_index=True)
```

```python
df_final
```

```python
import matplotlib.pyplot as plt
plt.plot(df_final['Num_of_clusters'], df_final['Inertia_Value'], '-o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Visualization')
plt.xticks(df_final['Num_of_clusters'])
plt.show()
```

**Based on the analysis of the above elbow graph, we choose K = 3**

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
mergings = linkage(df2[:-1], method= 'ward')
dendrogram(mergings, leaf_rotation=90, leaf_font_size=6)
plt.axhline(y=5, color='black', linestyle= '--')
plt.title('Dendogram for Hierarchal Clustering')
plt.show()
```

**Based on the analysis of the above Dendrogram, we choose K = 3**

**Performing KMeans Clustering**

```python
#using KMeans for 4 clusters.
from sklearn.cluster import KMeans
Kmeans = KMeans(n_clusters =3)
kmeans = Kmeans.fit(df2)
Pred_Kmean = kmeans.labels_
print(Pred_Kmean)
```

```python
df2['Pred_Class'] = Pred_Kmean
df2
```

{**Performing Hierarchical Clustering**}

```python
from sklearn.cluster import AgglomerativeClustering
cluster_hier_2 = AgglomerativeClustering() 
label_hier = cluster_hier_2.fit_predict(df2)
label_hier
```

```python
df2['Hier_Pred'] = label_hier
df2
```

```python
# Calculating Scores based on the KMean Cluster Method
from sklearn.metrics import silhouette_score
kmeans_silscore = silhouette_score(df2, df2['Pred_Class'])
print('The Silhouette Score using Kmean is: {}'.format(kmeans_silscore))
```

```python
# Calculating Scores based on the Hierarchical Clustering Method
from sklearn.metrics import silhouette_score
Hier_silscore = silhouette_score(df2, df2['Hier_Pred'])
print('The Silhouette Score using Hierarchical is: {}'.format(Hier_silscore))
```

**Dunn Index:** It is computed as lowest intercluster distance (i.e., the smallest distance between any two cluster centriods) divided by the highest intracluster distance(i.e., the largest distance between any 2 points in any cluster) as below,

Dunn Index should be maximized.

```python
#Calculating Dunn Index for KMean Clustering Data
from sklearn.metrics import pairwise_distances
from validclust import dunn
dist_df_KMeans = pairwise_distances(df2)
dunn_ind_KMeans = dunn(dist_df_KMeans,  kmeans.labels_)
print('Dunn Index for KMeans Clustering Data: '+ str(dunn_ind_KMeans))
```

```python
#Calculating Dunn Index for KMean Hierarchical Clustering Data
from sklearn.metrics import pairwise_distances
from validclust import dunn
dist_df_Hierar = pairwise_distances(df2)
dunn_ind_hierar = dunn(dist_df_Hierar,  label_hier)
print('Dunn Index for Hierarchical Clustering Data: '+ str(dunn_ind_hierar))
```

```python
#CH-Coeffiecent for KMeans Clustering
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score

ch_coeff_Kmean = calinski_harabasz_score(df2, df2['Pred_Class'])
print('CH Coeefiecent for KMean: ' + str(ch_coeff_Kmean))
```
